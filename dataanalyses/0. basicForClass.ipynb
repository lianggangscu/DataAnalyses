### 3.markdown 入门

1. **什么是markdown?**

   一种轻量级的标记语言,同html一样,都是用来写文档的.
   它的语法十分简单,能够实现以文本的方式对文档进结构化的编辑。

2. **markdown 的语法**

    2.1 标题

       markdown 支持6级标题,分别用1-6个`#`表示,`#`后面要加一个空格.

    2.2 列表

        markdown 支持有序列表和无序列表,分别用`1.`和`-`,`+`,`*`表示;
        有序列表和无序列表可以相互嵌套.

    2.3 引用

    引用表示引用了其他内容,用`>`表示,可以嵌套.引用在视觉效果上也能实现缩进的效果。

    2.4 图片

        在markdown纯文本中可以插入图片，插入图片的语法如下：
        图片用`!`表示,`!`后面要加`[]`表示图片的标题,`()`表示图片的链接地址. 
        图片的链接地址可以是本地的,也可以是网络上的.

    2.5 代码

    在markdown中,代码的表示方法有三种,分别是:
    
    - 代码块
    代码块用` ``` `表示,` ``` `后面要加一个空格,` ``` `后面要加一个语言的名称,表示代码块中的代码是哪种语言的代码.
    
    - 行内代码

    行内代码用 \` 表示,\`后面要加一个空格,\`后面要加一个语言的名称,表示代码块中的代码是哪种语言的代码.

    2.6 表格
        
        表格用`|`表示,`|`后面要加一个空格.
        表格的第一行要加`---`表示表格的分隔线.用于表头和数据的分割
        表格中的每一行可以使用三种方式实现对齐：
        - `:---` 表示左对齐
        - `---: `表示右对齐
        - `:---: `表示居中对齐  

    2.7 链接

        链接用`[]`表示,`[]`后面要加一个空格,`[]`后面要加一个链接地址.
        链接地址可以是本地的,也可以是网络上的.
        链接可以是文本链接,也可以是图片链接.

    2.8 分割线

        可以使用`---`表示分割线,也可以使用`***`表示分割线,也可以使用`___`表示分割线.

    2.9 字体设置

        - 可以使用`*`表示斜体,也可以使用`_`表示斜体.
        - 可以使用`**`表示粗体,也可以使用`__`表示粗体.
        - 可以使用`***`表示粗斜体,也可以使用`___`表示粗斜体.
        - 可以使用`~~`表示删除线,也可以使用`~~`表示删除线.
        - 可以使用`==`表示高亮,也可以使用`==`表示高亮.
        - 可以使用`$\color{red}{内容}$`, 设置内容字体颜色。
        
    2.10 脚注

        脚注用`[^]`表示,`[^]`后面要加一个空格,`[^]`后面要加一个脚注的名称.
    
    2.11 目录
    目录用`[TOC]`表示,`[TOC]`后面要加一个空格.

    2.12 数学公式

    数学公式用`$$`表示,`$$`后面要加一个空格,`$$`后面要加一个数学公式.

    2.13 支持内嵌html
    markdown 支持内嵌html,可以在markdown中插入html代码
    <tag attribut="value">content</tag>
